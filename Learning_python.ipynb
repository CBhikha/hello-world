## Reading a file from google drive and plotting a spectrogram

```python

#import the pyplot and wavfile modules 
import matplotlib.pyplot as plot
from scipy.io import wavfile

# Read the wav file (mono)
samplingFrequency, signalData = wavfile.read('/content/drive/MyDrive/Jane Ok.wav')
#samplingFrequency, signalData = wavfile.read('/content/drive/MyDrive/C195.wav')
signalData = signalData[:,1]

# Plot the signal read from wav file
plot.subplot(211)
plot.title('Spectrogram of a wav file')

plot.plot(signalData,linewidth=0.5)
plot.xlabel('Sample')
plot.ylabel('Amplitude')

plot.subplot(212)
ss, fr, t, im = plot.specgram((signalData),Fs=samplingFrequency,cmap='inferno')
plot.xlabel('Time')
plot.ylabel('Frequency')
plot.ylim(0,3500)

plot.show()
```

### Show image of a 2d array with interp

https://matplotlib.org/stable/gallery/images_contours_and_fields/interpolation_methods.html

```python
import numpy
plot.imshow(numpy.log(ss),origin='lower',interpolation='sinc', aspect='auto',cmap='magma')
plot.ylim(0,50)
```

```python
lss=numpy.log(ss)
plot.rcParams['figure.figsize'] = [10, 5]
ax=plot.plot(lss[0:30,range(200,400,10)])
```

```python
lss=numpy.log(ss)
l = lss.shape
l=l[1]
env = [0] * l
for i in range(0, l, 1):
  x = lss[0:30,i]
  imx = np.where(x > 5.2)
  imx = imx[0]
  mx = imx[len(imx)-1]
  env[i] = mx

plot.plot(env)
```

```python
fig = plt.figure(1, figsize=(20,8))

ax2 = plt.subplot(1,1,1)

ax2.imshow(numpy.log(ss),origin='lower',interpolation='sinc', aspect='auto',cmap='magma')

#ax2.set_aspect(3, "datalim")

ax21 = ax2.twinx()
env=numpy.array(env)
ax21.plot(env, alpha=0.7)

a = (np.diff(ax21.get_ylim())[0]/np.diff(ax1.get_xlim())*ss.shape[1]/ss.shape[0])
ax21.set_aspect(1./a, "datalim")

plt.show()
```

## FFT

```python
#import the pyplot and wavfile modules 
import matplotlib.pyplot as plot
from scipy.io import wavfile
from scipy.fft import fft, fftfreq, rfft

# Read the wav file (mono)
samplingFrequency, signalData = wavfile.read('/content/drive/MyDrive/Jane Ok.wav')
tf = signalData[:,1]

ff = rfft(tf) # rfft Only return one symetrical alf of the fft
ff = abs(ff)
plot.plot(ff,linewidth = 0.5)
plot.xlim(0, 4000)
plot.rcParams['figure.figsize'] = [15, 10] # [width, height]
```

```python
from scipy.signal import savgol_filter
import time
from IPython import display

for i in range(5):
  # smooth signal
  y = savgol_filter(ff, 20*i+5, 3)   # (x, window_length, polyorder)

  #plot with delay
  plot.plot(y,linewidth = 1)
  plot.xlim(0, 4000)
  display.display(plot.gcf())
  display.clear_output(wait=True)
  time.sleep(1)
```
