## Plotting uses matplotlib library

https://matplotlib.org/stable/api/index.html

```python
import matplotlib.pyplot as plt
  
# x axis values
x = [1,2,3,4,5,6]
# corresponding y axis values
y = [2,4,1,5,2,6]
  
# plotting the points 
plt.plot(x, y, color='green', linestyle='dashed', linewidth = 1,
         marker='o', markerfacecolor='blue', markersize=12)
  
# setting x and y axis range
plt.ylim(1,8)
plt.xlim(1,8)
  
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
  
# giving a title to my graph
plt.title('Some cool customizations!')
  
# function to show the plot
plt.show()
```

## Reading a file from google drive and plotting a spectrogram

```python

#import the pyplot and wavfile modules 
import matplotlib.pyplot as plot
from scipy.io import wavfile

# Read the wav file (mono)
samplingFrequency, signalData = wavfile.read('/content/drive/MyDrive/Jane Ok.wav')
signalData = signalData[:,1]

# Plot the signal read from wav file
plot.subplot(211)
plot.title('Spectrogram of a wav file')

plot.plot(signalData,linewidth=0.5)
plot.xlabel('Sample')
plot.ylabel('Amplitude')

plot.subplot(212)
plot.specgram((signalData),Fs=samplingFrequency,cmap='inferno')
plot.xlabel('Time')
plot.ylabel('Frequency')
plot.ylim(0,3500)

plot.show()
```

## String print with number variable

```python
quantity = 3
itemno = 567
price = 49
myorder = "I want {} pieces of item number {} for {:.2f} dollars."
print(myorder.format(quantity, itemno, price))
myorder = "I want {0} pieces of item number {1} for {2:.2f} dollars, {0} pieces."
print(myorder.format(quantity, itemno, price))
```
